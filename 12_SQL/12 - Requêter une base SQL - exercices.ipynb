1) Télécharger le fichier de la BDD depuis la page kaggle : https://www.kaggle.com/benhamner/sf-bay-area-bike-share/home <br>
2) A l'aide de sqlite3, connectez-vous à database.sqlite <br>

```python
import sqlite3
path = 'database_tp.sqlite'
conn = sqlite3.connect(path) # on crée une connexion 
conn
```

3) Affichez le résumé des tables présentes dans la BDD <br>

```python
import pandas as pd
tables = pd.read_sql("""SELECT *
                        FROM sqlite_master
                        WHERE type='table';""", conn)
tables
```

4) Créez le dataframe 'trip' contenant toutes les données de la table trip <br>

```python
trip = pd.read_sql("""SELECT *
                        FROM trip;""", conn)
trip.head()
```

5) A l'aide de count() comptez le nombres de voyages présents dans la table trip <br>

```python
trip = pd.read_sql("""SELECT COUNT(*)
                        FROM trip;""", conn)
trip
```

6) A l'aide de AVG, caculez le temps moyen d'un voyage (duration). Pour plus de clareté, renomnez la colonne 'Moyenne_loc_minutes'. Le temps est en seconde. Passez le en minutes en divisant Moyenne_loc_minutes par 60. <br>

```python
trip = pd.read_sql("""SELECT AVG(duration)/60 AS Moyenne_loc_minutes
                        FROM trip;""", conn)
trip
```

7) Affichez Moyenne_loc_minutes et count() en fonction du type d'abonnement (subscription_type) à l'aide de GROUP BY

```python
trip = pd.read_sql("""SELECT subscription_type, AVG(duration)/60 AS Moyenne_loc_minutes, count(*) AS count
                        FROM trip
                        GROUP BY subscription_type;""", conn)
trip
```

8) Créez le dataframe 'station' contenant toutes les données de la table station <br>

```python
station = pd.read_sql("""SELECT *
                        FROM station;""", conn)
station.head()
```

9) On souhaite récupérer le nom de la ville de départ de chaque voyage. A l'aide de Inner Join, récupérer la colonne city de la table city et ajouter la à la table trip <br>

```python
trip = pd.read_sql("""SELECT station.city AS city_start,*
                        FROM trip
                        INNER JOIN station
                        ON trip.start_station_id = station.id;""", conn)
trip.head()
```

10) Affichez maintenant Moyenne_loc_minutes et count(*) en fonction de la ville de départ à l'aide de GROUP BY <br>

```python
trip = pd.read_sql("""SELECT station.city AS city_start, count(*) as count, AVG(duration)/60 as Moyenne_loc_minutes
                        FROM trip
                        INNER JOIN station
                        ON trip.start_station_id = station.id
                        GROUP BY (city_start);""", conn)
trip.head()
```

11 ) Faites le même exercice mais avec la ville d'arrivée <br>

```python
trip = pd.read_sql("""SELECT station.city AS city_end, count(*) as count, AVG(duration)/60 as Moyenne_loc_minutes
                        FROM trip
                        INNER JOIN station
                        ON trip.end_station_id = station.id
                        GROUP BY (city_end);""", conn)
trip.head()
```

12) Affichez ce même tableau mais uniquement pour les utilisateurs avec abonnement ('Subscriber' dans la colonne subscription_type)

```python
trip = pd.read_sql("""SELECT station.city AS city_end, count(*) as count, AVG(duration)/60 as Moyenne_loc_minutes
                        FROM trip
                        INNER JOIN station
                        ON trip.end_station_id == station.id
                        WHERE subscription_type == 'Subscriber'
                        GROUP BY (city_end);""", conn)
trip.head()
```
