<h1> 15 - La statistique descriptive</h1>

- Definition wikipedia : L'objectif de la statistique descriptive est de décrire, c'est-à-dire de résumer ou représenter, par des statistiques, les données disponibles quand elles sont nombreuses.

```python
import pandas as pd

data = pd.read_csv('DigiDB_digimonlist.csv')
data.head(5)
```

<h2>1- Les types de variables </h2>
    
- <b>Les variables catégorielles (qualitatives) </b>
<p> Il n'y a pas de notion d'échelle, c'est un attribut d'un critère<br>
Exemples : [homme,Femme], [première classe, deuxième classe, troisième classe], [Vrai,Faux]<p>

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
sns.countplot(x="Type", data=data)
```

```python
sns.countplot(x="Attribute", data=data)
```

- <b>Les variables quantitatives </b>

<p> A l'inverse, ces variables permettent de mesurer des quantités <br>
    Il en existe 2 types : Variables quantitatives <b>discrètes</b> et <b>continues</b> </p>

<b>Les variables quantitatives discrètes </b>
<p> C'est une variable avec un nombre fini de valeurs <br>
Exemples : nombre d'items, nombre de personnes <p>

```python
#les variables quantitatives discrètes
sns.distplot(list(data['Memory']))
```

```python
sns.distplot(list(data['Lv 50 HP']))
```

<b>Les variables quantitatives continues </b>
<p> C'est une variable avec une infinité de valeurs <br>

<h2>2- Les statistiques de base </h2>

<b>Description d'une variable continue </b> <br>

    - La moyenne: Somme des valeurs de la variable divisée par le nombre d'individus



```python
mean = data['Lv 50 HP'].mean()
print(mean)
sns.distplot(list(data['Lv 50 HP']))
plt.axvline(mean, color='r')
```

    - La Médiane: valeur centrale qui partage l'échantillon en 2 groupes de même effectif: 50 % au-dessus et 50 % en dessous

```python
median = data['Lv 50 HP'].median()
print(median)
sns.distplot(list(data['Lv 50 HP']))
plt.axvline(mean, color='r')
plt.axvline(median, color='g')
```

    -Le mode: C'est la valeur la plus fréquente dans le jeu de données

```python
mode = data['Lv 50 HP'].mode()
print(mode)
sns.distplot(list(data['Lv 50 HP']))
plt.axvline(mean, color='r')
plt.axvline(median, color='g')
plt.axvline(float(mode), color='y')
```

```python
mode = data['Lv 50 HP'].mode()
print(mode)
nbValDif = len(data['Lv 50 HP'].unique())
sns.distplot(list(data['Lv 50 HP']), bins=nbValDif)
plt.axvline(mean, color='r')
plt.axvline(median, color='g')
plt.axvline(float(mode), color='y')
```

```python
mode = data['Type'].mode()
print(mode)
sns.countplot(x="Type", data=data)
```

    - la variance: disparité dans notre jeu de données: moyenne arithmétique des carrés des écarts à la moyenne arithmétique observée




```python
variance = data['Lv 50 HP'].var()
variance
```

    - L'écart type: c'est la racine carrée de la variance



```python
ecartType = data['Lv 50 HP'].std()
ecartType
```

```python
variance**(1/2)
```

    - Les quantiles: Dans la même logique que la médiane les quartile vont couper la population en 4 parties contenant le même nombre d'observation



```python
ax = sns.boxplot(x="Type", y="Lv 50 HP", data=data)
plt.show()
```

<h2>3- La loi normale </h2>

- <b> Quelques exemples <b>

```python
sns.distplot(list(data['Lv 50 HP']))
```

Dans une expérience aléatoire, plus on augmente le nombre d'obeservations et plus la répartition va se lisser et se rapprocher de la fonction de gauss


<b><p>Distribution de la fonction de Gauss :</p> <b>



<b> Calculer le coéficient d'assymétrie (skewness) de sa distribution </b>

Fait référence à l'asymétrie des données. Lorsque les données sont concentrées sur le côté droit de l'histogramme, par exemple, nous disons qu'elles ont un skew négatif. Lorsque les données sont concentrées à gauche, on dit qu'elles ont un skew positif.

```python
sns.distplot(list(data['Lv 50 HP']))
data['Lv 50 HP'].skew()
```

<b> Calculer le coéficient d'applatissement d'une distribution : Kurtosis </b>

<p>Plus la valeur d'aplatissement est élevée, plus le pic est élevé. Plus la valeur de kustose est faible, plus la valeur est étalée.</p>



```python
sns.distplot(list(data['Lv 50 HP']))
data['Lv 50 HP'].kurtosis()
```
