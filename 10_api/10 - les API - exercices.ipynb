6) Mettre l'information 'code' dans les paramètre d'appel de l'API (params=..) puis le supprimer de l'URL.

```python
import requests
from pandas.io.json import json_normalize

url = 'https://restcountries.eu/rest/v2/alpha/'
parametres = {
    'codes': 'FRA',
}

data = requests.get(url, params=parametres)
data = json.loads(data.content)
dataSet = json_normalize(data)
dataSet
```

7) Nous allons construire une fonction permettant d'obtenir une information précise sur un pays. 
Le comportement attendue est le suivant : la fonction infoPays('FRA','capital') devra retourner 'Paris'

7-1) Déclarer la fonction infoPays qui prendra 2 paramètres : 'code', correspondant au code du pays et 'info' correspondant au contenu de la colonne à retourner (capital, monnaie...)

7-2) placer dans la fonction l'appel à l'API réalisé à l'étape 6. la variable 'code' de l'api prendra la valeur du paramètre 'code' de la fonction infoPays

7-3) Toujours dans la fonction créer ensuite la variable 'info' qui recupera la valeur de la colonne correspondant au paramètre 'info' de la fonction

7-4) la fonction doit retourner la variable info en guise de réponse.

```python
def infoPays(code,info):
    parameters = {
    'codes': code,
    }
    url = 'https://restcountries.eu/rest/v2/alpha/'
    data = requests.get(url, params=parameters)
    data = json.loads(data.content)
    dataSet = json_normalize(data)
    info = dataSet[info][0]
    return info 
```
