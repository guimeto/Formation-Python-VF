## Pagination

## Requête POST

```python
# payload = {"name":"test", "description":"ceci est la descroption du repository"}
# requests.post("https://api.github.com/users/repos", json = payLoad)
```

```python
import requests
payload ={"name": "api-scrping"}
response = requests.post("https://api.github.com/user/repos", json = payload, 
                       headers=headers)
status = response.status_code
print(status)
```

## Requête PATCH/PUT

```python
# patch pour modifier un objet de notre reposetory
# put changer complètement l objet
```

```python
payload = {"name": "api","description":"formation udemy"}
response = requests.patch("https://api.github.com/repos/guimeto/api-scrping", 
                        json = payload, headers=headers)
status = response.status_code
print(status)
```

## Requête Delete

```python
# requête permet de supprimer un objet du serveur 
```

```python
response = requests.delete("https://api.github.com/repos/guimeto/api",headers=headers)
status = response.status_code
print(status)
```

## Training

```python
# creation d un repository sur nontre github
import requests
payload ={"name": "training"}
response = requests.post("https://api.github.com/user/repos", json = payload, 
                       headers=headers)
status = response.status_code
print(status)
```

```python
# mise a jour et on le renomme
payload = {"name": "Mon repository","description":"Training API"}
response = requests.patch("https://api.github.com/repos/guimeto/training", 
                        json = payload, headers=headers)
status = response.status_code
print(status)
```

```python
# suppression de notre repository 
response = requests.delete("https://api.github.com/repos/guimeto/Mon-repository",headers=headers)
status = response.status_code
print(status)
```
