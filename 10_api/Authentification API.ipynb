## 2-  Autres points d'accès

On peut aller sur d'autres comptes github: 

```python
# compte huandu 
response = requests.get("https://api.github.com/users/huandu", 
                       headers=headers)
huandu = response.json()
print(huandu)
```

```python
response = requests.get("https://api.github.com/orgs/facebook",
                       headers=headers)
orgs = response.json()
print(orgs)
```

## 3- La  pagination

Pour contrer les requêtes qui retournent un grand nombre d'objets.  

```python
params = {"per_page": 50, "page":1}
response = requests.get("https://api.github.com/users/rakeshsukla53/starred", 
                       headers=headers, params = params)
page1_repos = response.json()
print(page1_repos)
```

## 4- Point d'accès User-Level

## 5- Requête POST pour créer un objet 

Créer un objet sur le serveur API. On va l'utiliser pour créer un repertoire sur notre github.


Pour cela on va utiliser un dictionnaire 

 payload = {"name":"test", "description":"ceci est la descroption du repository"}
 
 requests.post("https://api.github.com/users/repos", json = payLoad)

```python
import requests
payload ={"name": "api-test"}

response = requests.post("https://api.github.com/user/repos", json = payload, 
                       headers=headers)
status = response.status_code
print(status)
```

## 6- Requête PATCH/PUT pour mettre à jour  un objet


- patch pour modifier un objet de notre repertoire
- put changer complètement l objet

```python
payload = {"name": "api","description":"formation udemy"}

response = requests.patch("https://api.github.com/repos/guimeto/api-test", 
                        json = payload, headers=headers)

status = response.status_code

print(status)
```

## 7- Requête Delete pour supprimer un objet

- requête qui permet de supprimer un objet du serveur 

```python
response = requests.delete("https://api.github.com/repos/guimeto/api-test",headers=headers)
status = response.status_code
print(status)
```
